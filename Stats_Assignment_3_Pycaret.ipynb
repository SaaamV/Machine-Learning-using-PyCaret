```python
#importing necessary libraries
import sys
import pandas as pd
from pycaret.classification import *
from matplotlib import pyplot as plt
import os
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
df=pd.read_csv(list(uploaded.keys())[0])
df.drop(columns=['peptide_sequence'], inplace=True)
```

```python
setup(data=df,target='Target',silent=True,train_size=0.75) #tried making train size=0.8 but increased computational time
```

```python
compare_models()
og=pull()
og_a=og.Accuracy
```

```python
setup(data=df,target='Target',silent=True,train_size=0.75,normalize=True,normalize_method='zscore')
compare_models()
temp=pull()
norm_zscore_accuracy=temp.Accuracy
```

```python
setup(data=df,target='Target',silent=True,train_size=0.75,normalize=True,normalize_method='minmax')
compare_models()
temp=pull()
norm_minmax_accuracy=temp.Accuracy
```

```python
setup(data=df,target='Target',silent=True,train_size=0.75,normalize=True,normalize_method='maxabs')
compare_models()
temp=pull()
norm_maxabs_accuracy=temp.Accuracy
```

```python
setup(data=df,target='Target',silent=True,train_size=0.75,normalize=True,normalize_method='robust')
compare_models()
temp=pull()
norm_robust_accuracy=temp.Accuracy
```

```python
norm_dict={"without normalization":og_a,"zscore":norm_zscore_accuracy,"minmax":norm_minmax_accuracy,"maxabs":norm_maxabs_accuracy,"robust":norm_robust_accuracy}
norm=pd.DataFrame(norm_dict)
norm.columns=pd.MultiIndex.from_tuples([("Accuracy without Normalization",''),("Accuracy using Normalization","zscore"),("Accuracy using Normalization","minmax"),("Accuracy using Normalization","maxabs"),("Accuracy using Normalization","robust")])
norm.index.names=['Model']
norm.to_csv('output-101903793-Normalization.csv')
```

```python
featsel_dict={"without feature selection":og_a,"classic 0.2":featsel_classic_2,"classic 0.5":featsel_classic_5,"boruta 0.2":featsel_boruta_2,"boruta 0.5":featsel_boruta_5}
featsel=pd.DataFrame(featsel_dict)
featsel.columns=pd.MultiIndex.from_tuples([("Accuracy without Feature Selection",''),("Accuracy using Feature Selection","Classic = 0.2"),("Accuracy using Feature Selection","Classic = 0.5"),("Accuracy using Feature Selection","Boruta = 0.2"),("Accuracy using Feature Selection","Boruta = 0.5")])
featsel.index.names=['Model']
featsel.to_csv('output-101903793-FeatureSelection.csv')
```
